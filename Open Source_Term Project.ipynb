```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
marketing = pd.read_csv('marketing.csv',sep=',')
income = marketing['income']
meat = marketing['meat']
wine = marketing['wines']
plt.scatter(income, wine, color='red',label='wine',marker='o',alpha=.3)
plt.scatter(income, meat, color='blue',label='meat',marker='o',alpha=.5)
plt.xlabel('income',fontsize=15)
plt.ylabel('meat & wines',fontsize=15)
plt.title('The relationship between income, meat and alcohol')
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
marketing = pd.read_csv('marketing.csv',sep=',')
X = marketing["meat"]
y = marketing["wines"]
plt.plot(X, y, 'o')
plt.show()
```

```python
line_fitter = LinearRegression()
line_fitter.fit(X.values.reshape(-1,1), y)
line_fitter.predict([[500]])
line_fitter.coef_
line_fitter.intercept_
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.show()
```
